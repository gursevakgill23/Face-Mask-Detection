```python
############################################### Import Packages ##################################################
import tkinter
import os
from tkinter import *
from tkinter.messagebox import *
from tkinter.filedialog import *
import cv2
import numpy as np
from keras.models import load_model
import cv2
import numpy as np
import pyttsx3
import speech_recognition as sr
from pydub import AudioSegment
from pydub.playback import play
from PIL import ImageTk,Image


################################################################## Creating GUI Window #############################################################
window_size = "700x700"
root=Tk()
class MaskDetection(Frame):

    def __init__(self, master=None):
        Frame.__init__(self, master)
        self.master = master

        ######################################### Sets default bg for top level windows ##########################################################
        self.tl_bg = "#EEEEEE"
        self.tl_bg2 = "#EEEEEE"
        self.tl_fg = "#000000"
        self.font = "Verdana 10"

        menu = Menu(self.master)
        self.master.config(menu=menu, bd=5)
# Menu bar

    # File
        file = Menu(menu, tearoff=0)
        menu.add_cascade(label="Start", menu=file)
        file.add_command(label="Detect the Mask", command=self.mask_detection)
        help_option = Menu(menu, tearoff=0)
        menu.add_cascade(label="Help", menu=help_option)
        #help_option.add_command(label="Features", command=self.features_msg)
        help_option.add_command(label="About this Project", command=self.msg)
        help_option.add_command(label="Develpoers", command=self.about)
########################################################## Mask Detection Logic ########################################################################
    def msg(self):
        tkinter.messagebox.showinfo("Face Mask detection","This project detects whether a person has weared mask or not,using tensorflow,opencv and CNN Deep Learning Algorithms  ")

    def about(self):
        tkinter.messagebox.showinfo("This Project Developers:","Gursevak Singh\nTkinter Documentation\nDeep Learning Course(Coursera)\nGithub Experts")

    def mask_detection(self):


        ################################ Function for coices ####################################################
        song1=AudioSegment.from_wav('Mask.wav')
        song2=AudioSegment.from_wav('No Mask.wav')

        engine = pyttsx3.init('sapi5')
        voices = engine.getProperty('voices')
        # print(voices[1].id)
        engine.setProperty('voice', voices[0].id)


        ############################################################## Function for voice ###########################################
        def speak(audio):
            engine.say(audio)
            engine.runAndWait()


        ################################################################# Lod Mdel ################################################################
        model = load_model('model-003.model')

        #################################################################  XML File for Face Detectoion #####################################################
        face_clsfr=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

        #################################################################### Capture webcam live stream #####################################################
        cap=cv2.VideoCapture(0)

        ############################################################# Setting labels and color of rectangle to show #############################################
        labels_dict={1:'MASK',0:'NO MASK'}
        color_dict={1:(0,255,0),0:(0,0,255)}

        ########################## While loop for capturing people with and without mask in webcam #########################################################
        while(True):

            ret,frame=cap.read()
            gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            faces=face_clsfr.detectMultiScale(gray,1.3,5)
        ################################################ Capturing people's faces on webcam ######################################################
            for (x,y,w,h) in faces:

                face_img=gray[y:y+w,x:x+w]
                resized=cv2.resize(face_img,(100,100))
                normalized=resized/255.0
                reshaped=np.reshape(normalized,(1,100,100,1))
                result=model.predict(reshaped) ## Predict the output from model(1 for with mask and 0 for without mask)

                label=np.argmax(result,axis=1)[0]

                cv2.rectangle(frame,(x,y),(x+w,y+h),color_dict[label],4)
                cv2.rectangle(frame,(x,y-40),(x+w,y),color_dict[label],4)
                cv2.putText(frame, labels_dict[label], (x, y-10),cv2.FONT_ITALIC, 1,(255,255,255),4)

                #if(labels_dict[label] =='MASK'):
                    #play(song1)
                #elif(labels_dict[label] =='NO MASK'):
                    #play(song2)
            cv2.imshow('Mask Detection System by Gursevak Singh',frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        cap.release()
        cv2.destroyAllWindows()


a = MaskDetection(root)
root.title('Mask Detection')
root.geometry(window_size)
txt_input=StringVar()
operator=''

Tops=Frame(root,width=100,height=100,relief=SUNKEN)
Tops.pack(side=TOP)
lblInfo=Label(Tops,font=('arial',28,'bold'),text='FACE MASK DETECTION SYSTEM\nBy Gursevak Singh\n',fg='Blue',bd=10,anchor='w')
lblInfo.grid(row=1,column=0)
canvas = Canvas(root, width = 600, height = 600)
canvas.pack()
img = ImageTk.PhotoImage(Image.open("guri1.jpg"))
canvas.create_image(10, 10, anchor=NW, image=img)
root.mainloop()
```
