```python
############################################### Import Packages ##################################################
import os
import cv2
import numpy as np
from keras.models import load_model
import cv2
import numpy as np
################################################################# Lod Mdel ################################################################
model1 = load_model('model-007.model')
# model2 = load_model('gendermodel-009.model')
#################################################################  XML File for Face Detectoion #####################################################
face_clsfr=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#################################################################### Capture webcam live stream #####################################################
cap=cv2.VideoCapture(0)
############################################################# Setting labels and color of rectangle to show #############################################
labels_dict={1:'MASK',0:'NO MASK'}
# gender_dict={0:'Man',1:'Women'}
color_dict={1:(0,255,0),0:(0,0,255)}
while(True):
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_clsfr.detectMultiScale(gray,1.3,5)
################################################ Capturing people's faces on webcam ######################################################
    for (x,y,w,h) in faces:
        face_img=gray[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(100,100))
        normalized=resized/255.0
        reshaped=np.reshape(normalized,(1,100,100,1))
        result1=model1.predict(reshaped)## Predict the output from model(1 for with mask and 0 for without mask)
#         result2=model2.predict(reshaped)
        label1=np.argmax(result1,axis=1)[0]
#         label2=np.argmax(result2,axis=1)[0]
        cv2.rectangle(frame,(x,y),(x+w,y+h),color_dict[label1],4)
        cv2.rectangle(frame,(x,y-40),(x+w,y),color_dict[label1],4)
        cv2.putText(frame, labels_dict[label1], (x, y-10),cv2.FONT_ITALIC, 1,(255,255,255),4)
#         cv2.putText(frame, gender_dict[label2], (x, y-50),cv2.FONT_ITALIC, 1,(0,0,255),4)

    cv2.imshow('Mask Detection System by Gursevak Singh',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()


```
